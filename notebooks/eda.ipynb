```python
# Our train data has no null data points
# Getting our X and y data
y = train_data['diagnosed_diabetes']
X = train_data.drop(columns='diagnosed_diabetes')
#Getting numerical and categorical columns

cat_cols = [col for col in X.columns if(X[col].dtype == 'object')]

num_cols = [col for col in X.columns if(X[col].dtype == 'int64' or X[col].dtype == 'float64')]
print(cat_cols)
print(num_cols)
```

```python
# Visualising our data
for col in num_cols:
    sns.histplot(x=col, data=X)
    plt.xticks(rotation=90)
    plt.show()
```

```python
# Visualising our data
for col in cat_cols:
    sns.countplot(x=col, data=X)
    plt.xticks(rotation=90)
    plt.show()
```

```python
# Visualising our data
for col in num_cols:
    sns.boxplot(x=col, data=X)
    plt.xticks(rotation=90)
    plt.show()
```

```python
# Checking the corelation of the numerical features with target(y)

num_corr_target = X[num_cols].corrwith(y)
print(num_corr_target)

sns.boxplot(x=y, y=X['physical_activity_minutes_per_week'])
plt.xticks(rotation=90)
plt.tight_layout()
plt.show()
```
